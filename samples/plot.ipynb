```python
import matplotlib
import matplotlib.pyplot as plt


import numpy as np
```

```python
# Load data
x = [i+1 for i in range(1000)]
y1 = np.loadtxt('data_normal.txt')
y2 = np.loadtxt('data_random.txt')
y3 = np.loadtxt('data_fix_epi.txt')
y4 = np.loadtxt('data_bandits.txt')
```

```python
# Print performance ratio
print('1-A2C/Random Strategy (episode 1000)= {:.2%}'.format(1-y1[len(y1)-1]/y2[len(y2)-1]))
print('1-A2C/Fixed-within-each-episode (episode 1000)= {:.2%}'.format(1-y1[len(y1)-1]/y3[len(y3)-1]))
print('1-A2C/Multi-armed bandits (episode 1000) = {:.2%}'.format(1-y1[len(y1)-1]/y4[len(y4)-1]))
print('1-A2C/Multi-armed bandits (episode 200) = {:.2%}'.format(1-y1[200-1]/y4[200-1]))
print('1-A2C/Multi-armed bandits (episode 100) = {:.2%}'.format(1-y1[100-1]/y4[100-1]))
```

```python
# Plot
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.xlabel('Episode')
plt.ylabel('Cumulative Flow Completion Time')
plt.title('Cumulative Flow Completion Time Comparison')
plt.legend(['A2C', 'Random strategy', 'Fixed-within-each-episode','Multi-armed bandits'])
plt.savefig('flow_time_4lines.eps', format='eps', bbox_inches = "tight")
plt.show()
```
